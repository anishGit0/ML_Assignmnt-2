  # Calculate the IQR
IQR = Q3 - Q1

  # Define the bounds for outliers

  upper_bound = Q3 + 1.5 * IQR

  # Identify outliers
outliers = data[(data['km_driven'] < lower_bound) | (data['km_driven'] > upper_bound)]

  # Remove outliers from the dataset
data_cleaned = data[(data['km_driven'] >= lower_bound) & (data['km_driven'] <= upper_bound)]

  # Print the results
print(f"Number of outliers identified: {outliers.shape[0]}")
  
  print("Outliers:")

  print(outliers)

  print("\nCleaned dataset shape:", data_cleaned.shape)

8. Factors Influencing Average Car Prices by Engine Size

Performance and Features:

Cars with larger engines often offer better performance, including faster acceleration and higher top speeds.
These vehicles may come equipped with more advanced technology and luxury features, contributing to higher prices.

Manufacturers may pass these costs onto consumers, resulting in higher prices for vehicles with bigger engines.
Market Demand:

There is often a higher demand for vehicles with larger engines, especially in markets where performance is prioritized.

Fuel Efficiency Considerations:

However, advancements in engine technology have led to more efficient larger engines, which can mitigate concerns about fuel costs.
Insurance and Maintenance Costs:

Vehicles with larger engines often come with higher insurance premiums and maintenance costs, which can influence overall ownership expenses.
Buyers may factor these costs into their purchasing decisions, impacting the average price they are willing to pay.

9. Income Bracket Guidelines for Car Purchases

1/10th Rule:

A common guideline is the 1/10th rule, which suggests that individuals should spend no more than 10% of their gross annual income on the purchase price of a car.
For example, if a person earns $42,000 annually, they should limit their car budget to $4,200.
Median Income Examples:

With a median household income of about $62,000, the recommended spending cap would be around $6,200.
This rule helps prevent buyers from overextending their finances and incurring debt.
Impact of Income on Car Choices

Higher Income Brackets:

Individuals in higher income brackets may opt for luxury vehicles or higher-end models, often exceeding the 10% guideline.
They may also afford additional costs such as insurance, maintenance, and financing without significant financial strain.
Lower Income Brackets:

Those in lower income brackets are typically more constrained and may prioritize affordability and practicality in their vehicle choices.
They are more likely to adhere to the 1/10th rule to avoid financial stress and ensure they can cover other living expenses.
Consequences of Exceeding Budget

Financial Strain:

Spending more than the recommended percentage can lead to financial difficulties, including high monthly payments that consume a large portion of net income.
This can result in stress and reduced ability to save for other important financial goals, such as retirement or education.
Opportunity Costs:

Investing in a car that exceeds budget limits can prevent individuals from allocating funds to more productive investments, such as real estate or stocks.
The long-term financial impact of such decisions can be significant, as cars depreciate quickly and do not generate income.

12. Seasonal Trends in Car Sales

Spring Surge:

March and April are peak months for car sales, largely driven by tax refund season and improved weather conditions.
Dealerships often see a significant increase in customer traffic during this time, as consumers are more inclined to make purchases.
Fall Peak:

The fall season also experiences a surge in sales, primarily due to the release of new models by manufacturers.
This period typically sees a spike in sales leading up to the end of the year, as consumers look to take advantage of year-end deals and promotions.
Winter Sales:

Contrary to traditional trends, winter has seen stronger sales in recent years, attributed to aggressive discounts and holiday promotions.
Dealerships often clear out inventory during this time, making it an attractive period for buyers looking for deals.
Inventory Management:

Understanding these seasonal patterns allows dealerships to adjust their inventory and marketing strategies accordingly.
For instance, stocking up on SUVs and all-wheel-drive vehicles before winter or promoting convertibles in the summer can align with consumer demand.
Promotional Strategies:

Seasonal discounts and themed sales events can effectively attract customers.
Tailoring marketing campaigns to reflect seasonal themes can enhance engagement and drive sales.
